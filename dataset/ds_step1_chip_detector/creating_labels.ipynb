```python
# Convert annotations to XML

import json
import os
import xmltodict

JSON_PATH = './full_size/mixed/ann_json/'
OUTPUT_PATH = './resized_x10/2/ann_xml/'

    
    
def convert_ann(filename):
    #filename = 'FULL-2019-04-26-153332.png.json'

    #Read JSON data into the datastore variable
    if filename:
        true_filename = filename[:-5]
        with open(JSON_PATH + filename, 'r') as f:
            x = json.load(f)

    y = """{ "annotation" : {  "folder": "", "filename":"","path":"","source": { "database":"Unknown"},"size": { "width": 0, "height": 0, "depth": 0},"segmented" :0,"object": { "name":"", "pose": "Unspecified", "truncated" : 0, "difficult": 0 , "bndbox":{"xmin":0,"ymin":0, "xmax":0, "ymax": 0}}}}"""
    y = json.loads(y)

    y['annotation']['folder'] = 'train'
    y['annotation']['filename'] = true_filename
    y['annotation']['path'] = OUTPUT_PATH
    y['annotation']['size']['height'] = x['size']['height']
    y['annotation']['size']['width'] = x['size']['width']
    y['annotation']['size']['depth'] = 3
    y['annotation']['object']['name'] = x['objects'][0]['classTitle']
    y['annotation']['object']['bndbox']['xmin'] = int(x['objects'][0]['points']['exterior'][0][0]/3648*300)
    y['annotation']['object']['bndbox']['ymin'] = int(x['objects'][0]['points']['exterior'][0][1]/5472*450)
    y['annotation']['object']['bndbox']['xmax'] = int(x['objects'][0]['points']['exterior'][1][0]/3648*300)
    y['annotation']['object']['bndbox']['ymax'] = int(x['objects'][0]['points']['exterior'][1][1]/5472*450)
    
    xmlString = xmltodict.unparse(y, pretty=True)

    print('\nXML output(output.xml):')
    print(xmlString)

    with open(OUTPUT_PATH+true_filename+'.xml', 'w') as f:
        f.write(xmlString)


files = os.listdir(JSON_PATH)
for file in files:
    convert_ann(file)
```

```python
# Convert annotations to CSV

import csv
import os
import json

JSON_PATH = './full_size/mixed/ann_json/'
IMAGE_PATH = '/home/abert/Documents/NumeriCube/eurosilicone/dataset/resized/mixed/img_val/'
files = os.listdir(IMAGE_PATH)

resize_shape = (300, 450)

with open('ann_val.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    
    for file in files:
        if file:
            true_filename = file
            with open(JSON_PATH + file+'.json', 'r') as f:
                x = json.load(f)


        shape = (x['size']['height'],x['size']['width'])
        path_to_image = IMAGE_PATH + true_filename

        class_name = x['objects'][0]['classTitle']
        x1 = int(x['objects'][0]['points']['exterior'][0][0]/shape[0]*resize_shape[0])
        y1 = int(x['objects'][0]['points']['exterior'][0][1]/shape[1]*resize_shape[1])
        x2 = int(x['objects'][0]['points']['exterior'][1][0]/shape[0]*resize_shape[0])
        y2 = int(x['objects'][0]['points']['exterior'][1][1]/shape[1]*resize_shape[1])
        
        spamwriter.writerow([path_to_image,x1,y1,x2,y2,class_name])
```
